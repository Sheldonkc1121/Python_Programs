```python
def char_count(txt1, txt2):
    count = 0
    for i in txt2:
        if i == txt1:
            count += 1
    return count        
```

```python
char_count("b", "big fat bubble")
```

```python
def count_unique(s1, s2):
    word = s1 + s2
    letters = set(word)
    return len(letters)
    
```
